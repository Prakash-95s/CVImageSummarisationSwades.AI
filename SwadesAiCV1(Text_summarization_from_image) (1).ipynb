##Model and API for text extraction and summarization
**Dino V2 model for Text extraction pytesseract for giving extracted tensor to a OCR text and a classification model res_net for object detection and classification and also a kmeans clustering to get dominant color


```python
from fastapi import FastAPI, Request, File, UploadFile, HTTPException
from fastapi.responses import JSONResponse, Response
import logging
from pydantic import BaseModel
from typing import List
import base64
from PIL import Image, ImageStat
from io import BytesIO
import torch
import timm
import torchvision.transforms as transforms
from fastapi import FastAPI, Request, Response
from starlette.middleware.base import BaseHTTPMiddleware
from starlette.middleware.cors import CORSMiddleware
from transformers import AutoImageProcessor,AutoModel, pipeline
import numpy as np
from sklearn.cluster import KMeans
import pytesseract
import io
import torch
import nest_asyncio
from pyngrok import ngrok
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import uvicorn
# Initialize FastAPI app
app = FastAPI()
# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


# Set up CORS middleware
# Custom middleware to set CORS headers
class CORSMiddlewareCustom(BaseHTTPMiddleware):
    async def dispatch(self, request: Request, call_next):
        response = await call_next(request)
        response.headers["Access-Control-Allow-Origin"] = "http://localhost:3001"
        response.headers["Access-Control-Allow-Headers"] = "Origin, X-Requested-With, Content-Type, Accept"
        return response

app.add_middleware(CORSMiddlewareCustom)
# using only CPUis available
device = device = torch.device('cpu')



## Load DINOv2 model and processor
# Initialize the DINO v2 model and processor
processor =  AutoImageProcessor.from_pretrained('facebook/dinov2-base')
model = AutoModel.from_pretrained('facebook/dinov2-base')
model.to(device)
model.eval()
# Initialize a pre-trained image classification model
classification_model = timm.create_model('resnet50', pretrained=True)
classification_model.eval()
classification_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Class labels for ImageNet since its not loading properly we are giving labels
#imagenet_labels = dict(enumerate(open('/content/drive/My Drive/imagenet-simple-labels.json').readlines()))#please insert your own text file path imagenet-simple-labels.json



class ImagePayload(BaseModel):
    images: List[str]

def get_image_brightness(image: Image.Image) -> str:
    stat = ImageStat.Stat(image)
    brightness = sum(stat.mean) / len(stat.mean)
    if brightness > 128:
        return "bright"
    else:
        return "dull"

def get_dominant_color(image: Image.Image) -> str:
    image = image.resize((100, 100))  # Resize for faster processing
    data = np.array(image).reshape((-1, 3))
    kmeans = KMeans(n_clusters=1, n_init=10).fit(data)  # Set n_init explicitly
    dominant_color = kmeans.cluster_centers_[0]
    return f"RGB({int(dominant_color[0])}, {int(dominant_color[1])}, {int(dominant_color[2])})"

def generate_text_context(image: Image.Image) -> str:
    # Convert the image to tensor for DINOv2
    inputs = processor(images=image, return_tensors="pt")
    outputs = model(**inputs)

    # Convert the image to tensor for classification
    classification_image = classification_transform(image).unsqueeze(0)
    with torch.no_grad():
        classification_outputs = classification_model(classification_image)

    # Get the top 5 class labels
    _, indices = torch.topk(classification_outputs, k=5)
    class_labels = [imagenet_labels[i] for i in indices[0].tolist()]

    # Extract feature vector mean for textual summary
    feature_vector_mean = outputs.last_hidden_state.mean().item()

    # Get image brightness
    brightness = get_image_brightness(image)

    # Get dominant color
    dominant_color = get_dominant_color(image)

    # OCR to extract text
    ocr_text = pytesseract.image_to_string(image)

    # Create a summary based on the extracted features and classification results
    feature_summary = (
        f"The image contains objects that are likely related to: {', '.join(class_labels)}. "
        f"The overall brightness of the image is {brightness}. "
        f"The dominant color in the image is {dominant_color}."
    )

    # Combine feature summary and OCR text
    combined_summary = f"Extracted Features:\n{feature_summary}\n\nExtracted Text:\n{ocr_text}"

    return combined_summary if combined_summary else "No features or text detected."

@app.post("/summarize")
async def summarize(payload: ImagePayload):
    summaries = []
    for base64_image in payload.images:
        try:
            # Decode the base64 string
            header, base64_data = base64_image.split(",", 1)
            image_data = base64.b64decode(base64_data)
            image = Image.open(BytesIO(image_data)).convert("RGB")

            # Process the image with the generate_text_context function
            summary = generate_text_context(image)
            summaries.append(summary)
        except Exception as e:
            raise HTTPException(status_code=400, detail=f"Failed to process image: {e}")

    return {"summaries": summaries}

# Handle OPTIONS requests
@app.options("/summarize")
async def options_summarize():
    return JSONResponse(content={}, status_code=200)

# Run FastAPI in Jupyter notebook
nest_asyncio.apply()
# Set your ngrok authentication token
ngrok.set_auth_token("2kCbF1jcDi32kqTgmFSa8EtbWTC_3q45XBPqL2Ei11JT35VDo")
ngrok_tunnel = ngrok.connect(8001)
print('Public URL:', ngrok_tunnel.public_url)
if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8001)
```
